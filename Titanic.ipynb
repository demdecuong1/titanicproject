```python
import numpy as np
from numpy.random import randn
import matplotlib as mtl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
def mfc(hello):
    age,sex = hello
    if age < 16:
        return 'child'
    else:
        return sex
    
titanic_df['person']=titanic_df[['Age','Sex']].apply(mfc,axis=1)
```
